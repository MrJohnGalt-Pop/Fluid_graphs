```python
import numpy as np
import scipy as sp
from scipy.integrate import solve_ivp,odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib
```

```python
#Coutte Newtonian
U1 = 10
U2 = 4
H = 5
y_list = np.linspace(0, H, 50)
uy = lambda y : (U1 - U2) * y / H + U2
u = [uy(y) for y in y_list]
```

```python
plt.plot(y_list,u)
plt.xlabel('Distance to bottom plate')
plt.ylabel('Velocity')
```
