### Q5

$\frac{\partial u}{\partial t} = \nu\frac{\partial^2 u}{\partial y^2}$

$u(y=0,t) = u(y=H,t) = Ucos(\omega T)$

sol:

$u = Re(f(y)exp(i \omega t))$

where $f(y) = Acosh(\lambda y)-Bsinh(\lambda y)$

$A=U, B=U(\frac{cosh(\lambda H)-1}{sinh(\lambda H)}), \lambda = (i+1)\sqrt{\frac{\omega}{2 \nu}}$

```python
H = 10
ω,ν = 1,1
λ = complex(1,1)*np.sqrt(ω/(2*ν))
A = U
B = U*(np.cosh(λ*H)-1)/np.sinh(λ*H)
y = np.linspace(0, H, 100)
def u_5(t):
    return np.real((A*np.cosh(λ*y) - B*np.sinh(λ*y))*np.exp([complex(0,ω*t)]))

plt.plot(u_5(0), y, label = 't=0')
plt.plot(u_5(np.pi/2), y, label = 't=pi/2')
plt.plot(u_5(np.pi), y, label = 't=pi')
plt.plot(u_5(3*np.pi/2), y, label = 't=3pi/2')
plt.legend()
```
