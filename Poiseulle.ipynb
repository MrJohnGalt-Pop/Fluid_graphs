```python
#import matplotlib as mpl
#mpl.use('pgf')
import matplotlib.pyplot as plt
import numpy as np
#import matplotlib as mpl
```

```python
n=1
u, H = 1,1
y = np.arange(-H,H ,.01)
u_1 = (-1/abs(-1))*(n/(n+1))*(1/u*abs(-1))**(1/n)*(abs(y)**((n+1)/n)-H**((n+1)/n))
u_5 = (-5/abs(-5))*(n/(n+1))*(1/u*abs(-5))**(1/n)*(abs(y)**((n+1)/n)-H**((n+1)/n))
u_10 = (-10/abs(-10))*(n/(n+1))*(1/u*abs(-10))**(1/n)*(abs(y)**((n+1)/n)-H**((n+1)/n))
plt.plot(u_1,y, label = 'dP/dx=-1')
plt.plot(u_5,y, label = 'dP/dx=-5')
plt.plot(u_10,y, label = 'dP/dx=-10')
plt.title('Poiseuille at n = 1 (Newtonian)')
plt.ylabel('y')
plt.xlabel('x')
plt.grid(True)
plt.legend()
#plt.savefig('Poiseuille at n=1.png', dpi=1000)
#plt.savefig('Poiseuille at n=1.pgf', format='pgf')
```

```python
G = 1
x = np.arange(a,7,.01)
h_1 = -(a**4***2)/(2*g*x**2)
h_2 = (1/(*g))*(-a**4***2*(/(2*x**2)-2*G***2/x**4))
plt.plot(x,h_1, label = 'Newtonian')
plt.plot(x,h_2, label = 'Oldroyd-B')
plt.plot(-x,h_1, color = 'C0')
plt.plot(-x,h_2, color = 'C1')
plt.plot([a,a],[-(a**4***2)/(2*g*a**2),(1/(*g))*(-a**4***2*(/(2*a**2)-2*G***2/a**4)) ], color = 'k')
plt.plot([-a,-a],[-(a**4***2)/(2*g*a**2),(1/(*g))*(-a**4***2*(/(2*a**2)-2*G***2/a**4)) ], color = 'k')
plt.title('Height profile for a rotating rod')
plt.legend()
plt.ylabel('height')
plt.xlabel('r')
#plt.savefig('h(r).png', dpi=1000)
```

```python
# fix z so f(z) a constant
z = -0.05
f = -*g*z 
x = np.arange(a,7,.01)
p_1 = -(*a**4***2)/(2*x**2) +f
p_2 = (-*a**4***2/(2*x**2)+(2*G***2*a**4***2/x**4)) +f
plt.plot(x,p_1, label = 'Newtonian')
plt.plot(x,p_2, label = 'Oldroyd-B')
plt.ylabel('pressure')
plt.xlabel('r')
plt.grid(True)
plt.title('Pressure profile for a rotating rod at z=-0.05')
plt.legend()
```
