```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
ρ = 1
Ω = 10
τ = 1
a = 1
g = 9.81
```

```python
G = 1
x = np.arange(a,7,.01)
h_1 = -(a**4*Ω**2)/(2*g*x**2)
h_2 = (1/(ρ*g))*(-a**4*Ω**2*(ρ/(2*x**2)-2*G*τ**2/x**4))
plt.plot(x,h_1, label = 'Newtonian')
plt.plot(x,h_2, label = 'Oldroyd-B')
plt.plot(-x,h_1, color = 'C0')
plt.plot(-x,h_2, color = 'C1')
plt.plot([a,a],[-(a**4*Ω**2)/(2*g*a**2),(1/(ρ*g))*(-a**4*Ω**2*(ρ/(2*a**2)-2*G*τ**2/a**4)) ], color = 'k')
plt.plot([-a,-a],[-(a**4*Ω**2)/(2*g*a**2),(1/(ρ*g))*(-a**4*Ω**2*(ρ/(2*a**2)-2*G*τ**2/a**4)) ], color = 'k')
plt.title('Height profile for a rotating rod')
plt.legend()
plt.ylabel('height')
plt.xlabel('r')
```
