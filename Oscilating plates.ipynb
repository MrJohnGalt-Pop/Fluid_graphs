Also pls mention that the real part of the effective viscousity, Im part of the effective viscosity and the ratio between them(see the code below)

```python
import numpy as np
import matplotlib.pyplot as plt

def u_5(t, H, U, GP, tau, omega):
    G = GP / tau
    c = eta
    lambda_ = np.sqrt(omega * (complex(0, 1) - omega * tau) / (G * tau + c + c * omega * tau * complex(0, 1)))
    A = U
    B = U * (np.cosh(lambda_ * H) - 1) / np.sinh(lambda_ * H)
    y = np.linspace(0, H, 100)
    return np.real((A * np.cosh(lambda_ * y) - B * np.sinh(lambda_ * y)) * np.exp(complex(0, omega * t)))


def u_5n(t, H, U, GP, tau, omega):
    G = GP / tau
    c = eta
    lambda_ = complex(1,1)*np.sqrt(omega/(2*eta))
    A = U
    B = U * (np.cosh(lambda_ * H) - 1) / np.sinh(lambda_ * H)
    y = np.linspace(0, H, 100)
    return np.real((A * np.cosh(lambda_ * y) - B * np.sinh(lambda_ * y)) * np.exp(complex(0, omega * t)))

H = 10
U = 1
GP = 10
tau = 2
omega = 1
eta = 1

t_values = [0, np.pi / (2 * omega), np.pi / omega, 3 * np.pi / (2 * omega)]

plt.figure(figsize=(8, 6))
for t in t_values:
    plt.plot(u_5(t, H, U, GP, tau, omega), np.linspace(0, H, 100), label=f't={t:.2f}')

plt.legend()
plt.xlabel('Velocity')
plt.ylabel('y')
plt.title('Velocity Profile for Viscoelastic Fluid (Oldroyd-B Model)')
plt.grid(True)
plt.show()
#3D Graphs
t_values_1 = np.linspace(0, 2*np.pi, 1000)
u = np.zeros((1000,100))
for i in range(1000):
    u[i,:] = u_5(t_values_1[i], H, U, GP, tau, omega)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
yy = np.linspace(0,H,100)
tt = np.linspace(0,2*np.pi,1000)
y_list, t_list = np.meshgrid(yy,tt)
ax.plot_surface(t_list, y_list, u)
ax.set_xlabel('time(s)')
ax.set_ylabel('distance to top plate(m)')
ax.set_zlabel('velocity(m/s)')
#ax.set_title('Velocity Profile for Carreau Fluid between 2-oscillating plates')
plt.show()
#ratio between real component and complex component
effective_viscosity = eta + GP / (1 + omega**2 * tau**2) - GP * omega * tau / (1 + omega**2 * tau**2)*complex(0,1)
print(effective_viscosity.real)
print(effective_viscosity.imag)
print(effective_viscosity.real/effective_viscosity.imag)
```
